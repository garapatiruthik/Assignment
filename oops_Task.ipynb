```python
print("1Q. Creaye a vehicle class with init method having instance variables as name_of_vehicle, max_speed and average_of_vehicle")
class vehicle:
    def __init__(self, name_of_vehicle, max_speed, average_of_vehicle):
        self.name_of_vehicle=name_of_vehicle
        self.max_speed=max_speed
        self.average_of_vehicle=average_of_vehicle
print('Ans. We made a class vehicle with the following data:')
car=vehicle("Benz",200,70)
print(car.name_of_vehicle)
print(car.max_speed)
print(car.average_of_vehicle)
print("\n2Q. Create a child class car from the vehicle class created in que1 ,Which will inherit the vehicle class. create a method named seating_capacity which takes capacity as an argument and returns name of the vehicle and its seating capacity.\nAns.The required data is as follows for a 6 seter car")
class child_car(vehicle):
     def seating_capacity(self,seats):
         return str(seats) +" seater car with name "+ self.name_of_vehicle
my_car=child_car("RR",150,60)
print(my_car.name_of_vehicle) 
print(my_car.max_speed)
print(my_car.average_of_vehicle)
print(my_car.seating_capacity(6))
print("\n3Q. What is multiple inheritance, write a python code to demonstrate multiple inheritance.\nAns.")
print('\nAns.Multiple inheritance is a feature in object-oriented programming languages that allows a class to inherit attributes and methods from multiple base classes. In Python, a class can inherit from multiple classes by listing them in the parentheses after the class name.\nNow I can access method 1,2 and 3 using implement as shown below:')
class Base():
    def method1(self):
        print('This is my base')
class primary(Base):
    def method2(self):
        print('This is method 2')
class secondary(primary):
    def method3(self):
        print('This is method 3')
implement=secondary()
implement.method1()
implement.method2()
implement.method3()
print('\n4Q.What are getter and setter in python, create a class and create a getter and setter method in this python')
print('Ans.In Python, getters and setters are methods used to retrieve and set the values of class attributes. They are also known as accessor and mutator methods, respectively. Getters are used to fetch the value of an attribute, while setters are used to modify or update the value of an attribute.\nIt can be understood by the below example:')
class Person:
    def __init__(self, name):
        self._name = name

    def get_name(self):
        return self._name

    def set_name(self, new_name):
        self._name = new_name

# Create an instance of the class
person = Person("Ruthik G")

# Use the getter to retrieve the name
print(person.get_name())

# Use the setter to change the name
person.set_name("Ruthik Garapati")

# Use the getter again to retrieve the updated name
print(person.get_name())

print('\n5Q.What is method overriding in python. write a python code to demonstrate the method overriding.\nAn example to find the areas of a square and a circle')
class shape():
    def area_of_shape(self):
        print('calculating the area')
class circle(shape):
    def area_of_shape(self):
        radius=5
        pi=3.14
        area= pi*(radius**2)
        return "Area of the circle is "+str(area)
class square(shape):
    def area_of_shape(self):
        side=4
        area=side**2
        return "Area of the square is "+str(area)
sq=square()
print(sq.area_of_shape())
cir=circle()
print(cir.area_of_shape())
```
