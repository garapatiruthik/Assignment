```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats

# Load the dataset
data = pd.read_csv('winequality-red.csv', delimiter=';')

# Display basic statistics
print(data.describe())

# Plot distribution of each feature
features = data.columns[:-1]  # Exclude the target variable 'quality'

plt.figure(figsize=(15, 10))
for i, feature in enumerate(features, 1):
    plt.subplot(3, 4, i)
    sns.histplot(data[feature], kde=True)
    plt.title(f'Distribution of {feature}')
plt.tight_layout()
plt.show()

# Check for normality using Shapiro-Wilk test
normality_results = {feature: stats.shapiro(data[feature])[1] for feature in features}
print(normality_results)

# Interpretation
non_normal_features = [feature for feature, p_value in normality_results.items() if p_value < 0.05]
print(f'Non-normal features: {non_normal_features}')

# Transformations (example with log transformation)
transformed_data = data.copy()
for feature in non_normal_features:
    transformed_data[feature] = transformed_data[feature].apply(lambda x: np.log1p(x))

# Plot transformed distributions
plt.figure(figsize=(15, 10))
for i, feature in enumerate(non_normal_features, 1):
    plt.subplot(3, 4, i)
    sns.histplot(transformed_data[feature], kde=True)
    plt.title(f'Transformed Distribution of {feature}')
plt.tight_layout()
plt.show()
```
