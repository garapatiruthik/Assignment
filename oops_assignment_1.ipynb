```python
print('1Q.Explain class and object with respect to object oriented programming. with suitable exapmle.\nAns.In object-oriented programming (OOP) with Python, a class is a blueprint or a template that defines the structure and behavior of objects. It serves as a blueprint for creating objects of a specific type. A class encapsulates data (attributes) and functions (methods) that operate on that data.\n Below is an example.')
class student():
    def __init__(self,course,name,year):
        self.course=course
        self.name=name
        self.year=year
details=student('cse','Ruthik',2023)
print(details.name)
print(details.course)
print(details.year)
print('\n2Q.Name the four pillars of oops.\nAns.Here are the names of the four pillars of object-oriented programming (OOP) in Python are:\nEncapsulation\nInheritance\nPolymorphism\nAbstraction')
print('\n3Q.Explain why the __init__() function is used give a suitable example.\nAns.The __init__() function in Python is a special method known as the constructor. It is automatically called when an object is created from a class. The primary purpose of the __init__() method is to initialize the attributes or properties of an object.An example for __init__() function is below to initialize the bio details as below:')
class bio():
    def __init__(self,name,age,occupation):
        self.name=name
        self.age=age
        self.occupation=occupation
    def display(self):
        print(self.name)
        print(self.age)
        print(self.occupation)
info=bio("Ruthik",17,"student")
info.display()
print("\n4Q.Why self is used in oops.\nAns.In object-oriented programming (OOP) in Python, self is used as a convention to refer to the instance of a class. It acts as a reference to the current object being manipulated or accessed.When defining methods within a class, including the __init__() constructor, the first parameter is always self. This parameter allows the methods to access and manipulate the object's attributes and methods within the class.")
print('\n5Q.What is inheritance. Give an example for each type of inheritance.')
print("\nAns.Inheritance is a fundamental concept in object-oriented programming (OOP) that allows classes to inherit attributes and methods from other classes. In Python, there are several types of inheritance:\nSingle Inheritance:Single inheritance occurs when a class inherits from a single base class. It is the simplest form of inheritance. Here's an example:")
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        print("The animal speaks.")

class Dog(Animal):
    def __init__(self, name):
        super().__init__(name)
    def speak(self):
        print("The dog barks.")
        
my_dog = Dog("husky")
my_dog.speak()
print("\nMultiple Inheritance:\nMultiple inheritance occurs when a class inherits from more than one base class. It allows a class to inherit attributes and methods from multiple classes. Here's an example:")
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        print("The animal speaks.")

class CanFly:
    def fly(self):
        print("The animal can fly.")

class Bird(Animal, CanFly):
    def __init__(self, name):
        Animal.__init__(self, name)
        CanFly.__init__(self)

my_bird = Bird("lily")
my_bird.speak()
my_bird.fly()  
print("\nMultilevel Inheritance:\nMultilevel inheritance occurs when a derived class inherits from another derived class. It creates a hierarchy of classes. Here's an example:")
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        print("The animal speaks.")

class Dog(Animal):
    def __init__(self, name):
        Animal.__init__(self, name)

    def bark(self):
        print("The dog barks.")

class Labrador(Dog):
    def __init__(self, name):
        Dog.__init__(self, name)

    def play(self):
        print("The Labrador plays fetch.")

my_labrador = Labrador("Harry")
my_labrador.speak()
my_labrador.bark()   
my_labrador.play()
```
